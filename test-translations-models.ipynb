```python
metric = evaluate.load("sacrebleu")
```

```python
predictions = [
    "And it's weird for me, and it's all happening to me a thousand kilometres away from any point where I live, and in the midst of death, I come out of my pocket of paper and a pencil of ink."
]
references = [
    [
        "Absurd as it might seem to me, a thousand miles from any human habitation and in danger of death, I took out of my pocket a sheet of paper and my fountain pen."
    ]
]
metric.compute(predictions=predictions, references=references)
```
